# scraping tables from all pages

```python
# importing packages
import os
import sys
import pandas as pd
import shutil
import re
from datetime import datetime, date
from random import randint
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options

driver= webdriver.Chrome(executable_path='/Users/isita/Downloads/chromedriver')

project_title=[]
country=[]
project_id = []
commitment_amount = []
status = []
Approval_date =[]
last_updated_date =[]
link = 'https://projects.worldbank.org/en/projects-operations/projects-list?os='

for i in range(1078) : 
    link = driver.get('https://projects.worldbank.org/en/projects-operations/projects-list?os='+str(i*20))
    sleep(2)
    elements = driver.find_elements_by_xpath('//table/tbody/tr')
    sleep(1)
    for i in elements:
        sub_elements = i.find_elements_by_tag_name('td')
        print(len(sub_elements))
#       print(sub_elements)
        if len(sub_elements)==7:
            # appending data into the lists
            project_title.append(sub_elements[0].text)
            country.append(sub_elements[1].text)
            project_id.append(sub_elements[2].text)
            commitment_amount.append(sub_elements[3].text)
            status.append(sub_elements[4].text)
            Approval_date.append(sub_elements[5].text)
            last_updated_date.append(sub_elements[6].text)
        for j in sub_elements:
            print(j.text)
        
    
```

# creating a dataframe from the scraped data

```python
df = pd.DataFrame({'project_title':project_title,
                       'Country':country,
                       'project_id': project_id,
                       'commitment_amount': commitment_amount,
                       'status': status,
                       'Approval_date': Approval_date,
                       'last_updated_date': last_updated_date})
```
