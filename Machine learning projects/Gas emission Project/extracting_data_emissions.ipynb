```python
from geopy.geocoders import Nominatim
# names of cities
cities = ['Tokyo','Jakarta','Delhi','Manila','Seoul',
          'Shanghai','Karachi','Beijing','New York','Guangzhou','Sao Paulo','Mexico City','Mumbai','Kyoto','Moscow',
          'Dhaka','Cairo','Los Angeles','Bangkok','Kolkata']
# lat and long
geolocator = Nominatim(user_agent="MyApp")
lat_long=[]

for i in cities:
    location = geolocator.geocode(i)
    lat_long.append((abs(location.latitude),abs(location.longitude)))
    
```

```python
import pandas as pd
import numpy as np


gases = ['methane','carbonmonoxide','ozone','nitrogendioxide']

for k in gases:
    interval_start = []
    max_time =[]
    min_time =[]
    average_value =[]
    minimum_value=[]

    for i in lat_long:
        link = "https://api.v2.emissions-api.org/api/v2/"+k+"/statistics.json?interval=day&begin=2021-05-20&end=2022-05-20&point="+str(i[1])+','+str(i[0])
        response = requests.get(link)
        data = response.text
        data = json.loads(data)
        l1=[]
        l2=[]
        l3=[]
        l4=[]
        l5=[]
        for i in data:
            l1.append(i['time']['interval_start'])
            l2.append(i['time']['max'])
            l3.append(i['time']['min'])
            l4.append(i['value']['average'])
            l5.append(i['value']['min'])
        interval_start.append(l1)
        max_time.append(l2)
        min_time.append(l3)
        average_value.append(l4)
        minimum_value.append(l5)
    main_data=[]
    for i,j  in enumerate(cities):
        dictionary = {j:{'interval_start':interval_start[i],
                         'max_time':max_time[i],
                         'min_time':min_time[i],
                         'average_value':average_value[i],
                         'minimum_value':minimum_value[i]}}
        reform = {(outerKey, innerKey): values for outerKey, innerDict in dictionary.items() for innerKey, values in innerDict.items()}
        main_data.append(pd.DataFrame(reform))
    
    data = pd.concat(main_data,axis=1)
    data.to_csv(k+'.csv')
```
