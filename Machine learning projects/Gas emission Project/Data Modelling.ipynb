```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# data modelling for carbon monoxide tokyo data

```python
df = pd.read_csv('carbonmonoxide.csv',header=[0,1])
```

```python
df.head()
```

```python
df['Tokyo'].loc[:166]
```

```python
data= df['Tokyo'].loc[:166]
data
```

```python
data.info()
```

```python
data['interval_start'] = pd.to_datetime(data['interval_start'])
data['max_time'] = pd.to_datetime(data['max_time'])
data['min_time'] = pd.to_datetime(data['min_time'])
```

```python
data.info()
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.histplot(data=data,x='average_value')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X= data[['average_value','minimum_value']]
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.scatterplot(data= data,x='interval_start',y='average_value',hue= cluster_labels)
plt.title('Tokyo carbonmonoxide average value data')
```

# data modelling for methane data kolkata

```python
df2 = pd.read_csv('methane.csv',header=[0,1])
```

```python
df2.head()
```

```python
data2 = df2['Kolkata']
```

```python
data2.head()
```

```python
data2['interval_start'] = pd.to_datetime(data2['interval_start'])
data2['max_time'] = pd.to_datetime(data2['max_time'])
data2['min_time'] = pd.to_datetime(data2['min_time'])
```

```python
data2.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X= data2[['average_value','minimum_value']]
scaled_X = scaler.fit_transform(X)
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
cluster_labels = model.fit_predict(scaled_X)
cluster_labels
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.scatterplot(data= data2,x='interval_start',y='average_value',hue= cluster_labels)
plt.title('Kolkata methane average value data')
```
