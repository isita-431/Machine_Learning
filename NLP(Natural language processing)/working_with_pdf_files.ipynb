## Appending one pdf to another

```python
pdf1= open('Business_Proposal.pdf', 'rb')
```

```python
reader = PyPDF2.PdfFileReader(pdf1)
```

```python
page_one = reader.getPage(0)
```

```python
pdf_writer = PyPDF2.PdfFileWriter()
```

```python
pdf_writer.addPage(page_one)
```

```python
pdf_output = open('new_pdf.pdf','wb')
```

```python
pdf_writer.write(pdf_output)
```

```python
pdf_output.close()
```

```python
pdf1.close()
```
