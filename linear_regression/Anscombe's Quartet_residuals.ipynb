```python
# importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# anscombes_quartet
Anscombe's quartet comprises of four quartets that have similar descriptive statistics but different graphical distributions. 
linear regression isn't the best model for all the quartets

```python
#anscombes_quartet1.csv
quartet = pd.read_csv('anscombes_quartet1.csv')
```

```python
quartet.head()
```

```python
quartet['pred_y'] = 4 + 0.5 * quartet['x']
quartet['residual'] = quartet['y'] - quartet['pred_y']
sns.scatterplot(data=quartet,x='x',y='y')
sns.lineplot(data=quartet,x='x',y='pred_y',color='red')
plt.vlines(quartet['x'],quartet['y'],quartet['pred_y'])
```

```python
sns.kdeplot(quartet['residual'])
```

the kde plot must be nearly normal. 

```python
sns.scatterplot(data=quartet,x='y',y='residual')
plt.axhline(y=0, color='r', linestyle='--')
```

in the residual plot above, the points are scattered randomly , without any pattern, so we can plot it with linear fit.

```python
# second quartet
quartet = pd.read_csv('anscombes_quartet2.csv')
```

```python
quartet.head()
```

```python

# y = 4.00 + 0.500x
quartet['pred_y'] = 4 + 0.5 * quartet['x']
quartet['residual'] = quartet['y'] - quartet['pred_y']

sns.scatterplot(data=quartet,x='x',y='y')
sns.lineplot(data=quartet,x='x',y='pred_y',color='red')
plt.vlines(quartet['x'],quartet['y'],quartet['pred_y'])
```

```python
sns.kdeplot(quartet['residual'])
```

```python
sns.scatterplot(data=quartet,x='y',y='residual')
plt.axhline(y=0, color='r', linestyle='--')
```

in the residual plot above , a pattern can be seen, and the no. of points above and below 0 differ too much in count, which says linear fit isn't the best way.
